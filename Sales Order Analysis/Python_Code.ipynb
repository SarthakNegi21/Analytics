```python
import pandas as pd
```

```python
#read the orders.csv file while converting "Not Available" and "unknown" values in Ship Mode columns to null

df = pd.read_csv("/Users/sarthaknegi/Documents/Python Programs/archive.zip",na_values=['Not Available', 'unknown'])
df.head(10)
```

```python
df["Ship Mode"].unique()
```

```python
df.columns
```

```python
#manipulating column names to lower case and adding a underscore in between their names

df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.head(10)
```

```python
#create additional columns for discount,selling price and profit

df["discount"] = df["list_price"] * (df["discount_percent"]/100)
df["sale_price"] = df["list_price"] - df["discount"]
df["profit"] = df["sale_price"] - df["cost_price"]
df
```

```python
#change datatype of order_date column to datetime

df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
#remove columns - cost_price,list_price,discount_percent

df.drop(["cost_price","list_price","discount_percent"],axis=1,inplace=True)
```
